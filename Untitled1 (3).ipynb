```python
import pandas as pd
import numpy as np
import re
```

```python
dfMales=pd.read_csv("/home/datascience/notebooks/Nikhil/Indian-Male-Names.csv")
dfFemales=pd.read_csv("/home/datascience//notebooks/Nikhil/Indian-Female-Names.csv")
```

```python
df=pd.concat([dfMales,dfFemales]).reset_index(drop=True)
```

```python
def remove_punctuations(text):
    rmvSpcchar1 = re.compile("[/(){}\[\]\|\-\t]")
    rmvSpcchar = re.compile("[/(){}\[\]\|&@,;_'.-]")
    rmvBadSym = re.compile('[^\d+\0-9a-z#+_]')
    rmvNumbers = re.compile('\d+')
    text = re.sub(rmvSpcchar1, ' ', text)
    text = re.sub(rmvSpcchar, ' ', text)
    text = re.sub(rmvBadSym, ' ', text)
    text = re.sub(rmvNumbers, ' ', text)
    return text

def is_vowel(char):

    if char in ('a', 'e', 'i', 'o', 'u'):
        return'1'
    else:
        return'0' 
    
    
```

```python
df=df[df["name"].notna()]
df.fillna(' ', inplace=True)
df
```

```python
df["Firstname"] = df["name"].apply(lambda x: remove_punctuations(x))
```

```python
from collections import OrderedDict
df['First_Name']=df['Firstname'].apply(lambda x: list(filter(None,OrderedDict.fromkeys(x.split(" ")))))
df
```

```python
df['Counts'] = df['First_Name'].apply(lambda x: len(x))
df
```

```python
df.drop(index_names, inplace = True)
df
```

```python
df['First_name'] = df['First_Name'].apply(lambda x: x[0])
df
```

```python
df['Last_Name'] = df['First_Name'].apply(lambda x: x[1] if len(x)>1 else "")
df
```

```python
df.drop(['First_Name','Firstname','name','race','Counts'],axis=1,inplace=True)
df
```

```python
df["Vowel_Present"] = df["First_name"].apply(lambda x: is_vowel(x[-1]))
```

```python
df["nameLen"] = df["First_name"].apply(lambda x: len(x))
dfClean = df[df["nameLen"]>=3]
```

```python
gender_count=dfClean[['gender','Vowel_Present']].value_counts()
gender_count
```

```python
Females = dfClean[dfClean["gender"]=="f"]
Females_0=Females[Females['Vowel_Present']=='0']
Females_0["lastLetter"] = Females_0["First_name"].apply(lambda x: x[-1])
fDf = Females_0["lastLetter"].value_counts().reset_index()
fDf["%occurance"] = fDf["lastLetter"]*100/fDf["lastLetter"].sum()
fDf
```

```python
males = dfClean[dfClean["gender"]=="m"]
Males_0 = males[males['Vowel_Present']=='1']
Males_0["lastLetter"] = Males_0["First_name"].apply(lambda x: x[-1])
mDf = Males_0["lastLetter"].value_counts().reset_index()
mDf["%occurance"] = mDf["lastLetter"]*100/mDf["lastLetter"].sum()
mDf
```
