```python
def name_lists(name_list):
    d = {}
    
    for name in name_list:
                
        first_name = name.split()[0]
        middle_name = name.split()[1]
        last_name = name.split()[2]
        if first_name in d:
            
            d[first_name].append(name)
            d[first_name].sort()
        elif middle_name in d:
            d[middle_name].append(name)
            d[middle_name].sort()
        elif last_name in d:
            d[last_name].append(name)
            d[last_name].sort()
            
        else:
            d[first_name]=[name]
            d[middle_name]=[name]
            d[last_name]=[name]

    return d
```
