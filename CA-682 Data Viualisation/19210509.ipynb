```python
import pandas as pd
%matplotlib inline

online_csv = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(online_csv)
pd.crosstab(df.Survived, df.Pclass,
            rownames=["Survived"],
            colnames=['PClass'],
            dropna=False)
```

```python
df[df.Survived==1].Survived.value_counts()
```

```python
df[df.Fare!='0'].Survived.value_counts().plot(kind='barh',x="Fare",y="Survived")
```
